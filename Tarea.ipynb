```python
import pandas as pd 
```

```python
adidas = pd.read_csv("adidas.csv")
```

```python
adidas.head(10)
```

```python
adidas.columns
```

```python
cols = ['url', 'name', 'original_price', 'currency',
       'availability', 'color', 'category', 'source', 'source_website',
       'breadcrumbs', 'description', 'brand', 'images', 'country', 'language', 'crawled_at']
```

```python
cols_validas = []
cols_many_levels = []
cols_one_level = []

size_df = adidas.shape[0]
umbral = 0.98
```

```python
for c_ in cols: 
    nivel = adidas.loc[:, c_].drop_duplicates().shape[0]
    if (nivel/size_df) >= umbral: 
        cols_many_levels.append(c_)
    elif nivel == 1: 
        cols_one_level.append(c_)
    else: 
        cols_validas.append(c_)
```

```python
adidas.loc[:,["category"]].value_counts()
```

```python
levels = adidas.groupby(["category"],as_index=False).agg({"sku": ["count"]})
```

```python
levels.columns = levels.columns.droplevel(1)
```

```python
levels
```

```python
levels["percent"] = levels["sku"]/size_df
levels.sort_values(["percent"], ascending = False)
```

```python
levels.sort_values(["percent"], ascending = False).shape
```

```python
levels["category_t"] = "Others"
```

```python
levels
```

```python
levels["category_t"] = levels.apply(lambda j: j["category"] if j["percent"] > 0.3 else j["category_t"], axis=1)
```

```python
levels.loc[:, ["category", "category_t"]]
```

```python
adidas.merge(levels, on=["category"], how = "inner")
```
